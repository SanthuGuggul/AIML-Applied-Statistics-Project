Pythagorean Expectation is a formula used in basketball to estimate a team's winning percentage based on their points scored and points allowed. It was originally developed by Bill James for baseball, but it has been adapted for basketball as well. The formula is as follows:

Pythagorean Expectation = (Points Scored)^n / ((Points Scored)^n + (Points Allowed)^n)

In this formula, "Points Scored" represents the total number of points scored by the team throughout the season, and "Points Allowed" represents the total number of points allowed by the team's opponents. The value of "n" is usually set to 13, which is an empirically derived constant.
